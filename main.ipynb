```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("./data/rides.csv")
```

```python
print(data.head())
```

```python
data.info()
```

As, we can see there are some null values in the Rides completed columns.
Let's see the number of null values and drop it since we cannot recover the null values.

```python
data.isnull().sum().reset_index()
```

Here, Rides Completed column has 54 null values.

##### Dropping the null values:

```python
data = data.dropna()
```

```python
data.isnull().sum().reset_index()
```

#### Demand and Supply Analysis:

```python
demand = data['Riders Active Per Hour']
supply = data['Drivers Active Per Hour']

fig = px.scatter(data, x=supply, y=demand, trendline='ols', title='Demand and Supply for the Ride')
fig.update_layout(xaxis_title='Number of Drivers active per hour(Supply)',
                 yaxis_title = 'Number of Riders Active per hour(Demand)')
fig.show()
```

Here, the relationship betweent the driver and riders available is consistent.
